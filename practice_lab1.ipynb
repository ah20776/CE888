```
import matplotlib.pyplot as plt
x1=[1,2,3]
y1=[2,4,1]
plt.plot (x1,y1,label="line 1")

x2=[1,2,3]
y2=[4,1,3]
plt.plot (x2,y2,label="line 2")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Two lines in same graph")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,4,1,5,2,6]
plt.plot(x,y,color="green",linestyle="dashed",linewidth=3,marker="o",markerfacecolor="blue",markersize=12)

plt.ylim(0,8)
plt.xlim(0,8)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("example 1")
plt.show()
```

```
from numpy import sin
import matplotlib.pyplot as plt

x=[x*0.1 for x in range(100)]
y=sin(x)

plt.plot(x,y)
plt.show()
```
